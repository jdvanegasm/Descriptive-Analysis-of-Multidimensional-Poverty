```python
#Funcion para calcular el coeficiente de Gini
def coeficiente_gini(x):
    sorted_x = np.sort(x)
    n = len(x)
    cumx = np.cumsum(sorted_x)
    gini = (n + 1 - 2 * (np.sum(cumx) / cumx[-1])) / n
    return gini
```

```python
# Diccionario para almacenar los coeficientes de Gini
gini_resultados = {}

# Calcular el coeficiente de Gini para cada departamento
for nombre, codigo in departamentos_representativos.items():
    data_filtrada = data[data['dpto'] == codigo]
    ingresos = data_filtrada['inglabo'].dropna().values
    gini_resultados[nombre] = coeficiente_gini(ingresos)

# Mostrar los resultados del coeficiente de Gini para cada departamento
for nombre, gini in gini_resultados.items():
    print(f"Coeficiente de Gini para {nombre}: {gini}")

# Crear un DataFrame a partir de los resultados del coeficiente de Gini
df_gini_resultados = pd.DataFrame(list(gini_resultados.items()), columns=['Departamento', 'Gini'])
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(14, 8))
scatterplot = sns.scatterplot(x='Gini', y='Departamento', data=df_gini_resultados, s=100, color='b', marker='o')

for index, row in df_gini_resultados.iterrows():
    scatterplot.text(row.Gini + 0.002, row.Departamento, round(row.Gini, 4), color='black', ha="center", va='center')

plt.title('Coeficiente de Gini para Diferentes Departamentos', fontsize=16)
plt.xlabel('Coeficiente de Gini', fontsize=14)
plt.ylabel('Departamentos', fontsize=14)
plt.grid(axis='x')
plt.show()
```


Para evaluar la pobreza multidimensional en cada departamento, se establecerán umbrales específicos basados en una serie de indicadores clave que reflejan diversas dimensiones de la pobreza. Los indicadores seleccionados y sus respectivos umbrales de privación se describen de la siguiente manera:

**Material de Vivienda (p4010):** Este indicador identifica la calidad del material de la vivienda. Se asignará un valor de 1, indicativo de privación, si el material de la vivienda incluye alguno de los siguientes materiales menos duraderos o provisionales: adobe, tapia pisada, zinc, tela, cartón, latas, desechos, o plástico. De lo contrario, se asignará un 0, indicando la ausencia de privación.

**Promedio de Años de Educación (p6210s1):** La educación es un indicador crítico del bienestar socioeconómico. Para este indicador, el umbral de privación se establece en menos de 6 años de educación. Si un individuo ha recibido menos de 6 años de educación, se considera en situación de privación y se asigna un valor de 1; si no, se asigna un 0.

**Ingresos Laborales (inglabo):** Este indicador mide la suficiencia de los ingresos laborales con respecto a un umbral mínimo establecido. Un ingreso laboral inferior a 250,000 pesos será considerado como un indicativo de privación económica, y se asignará un valor de 1 a este indicador; de lo contrario, se asignará un 0.

**Afiliación a la Salud (p6090):** La cobertura de salud es esencial para la seguridad y el bienestar. En este caso, la falta de afiliación a un sistema de salud se considera una privación, asignando un valor de 1 al indicador si la persona no está afiliada a ningún sistema de salud; si está afiliada, se asigna un 0.

Cada uno de estos indicadores se cuantificará y estandarizará para cada registro en los datasets correspondientes a los diferentes departamentos. La suma de estos valores para cada individuo ofrecerá una medida compuesta que reflejará el nivel de privación multidimensional.

```python
umbrales = {
    'p4010': 1,  # 1 si incluye materiales menos duraderos, 0 de lo contrario
    'p6210s1': 1,  # 1 si menos de 6 años de educación, 0 si 6 años o más
    'inglabo': 1,  # 1 si ingresos menores a 250,000 pesos, 0 de lo contrario
    'p6090': 1  # 1 si no está afiliado a la salud, 0 si está afiliado
}

ponderadores = {
    'p4010': 1,  # 1 si incluye materiales menos duraderos, 0 de lo contrario
    'p6210s1': 1,  # 1 si menos de 6 años de educación, 0 si 6 años o más
    'inglabo': 1,  # 1 si ingresos menores a 250,000 pesos, 0 de lo contrario
    'p6090': 1  # 1 si no está afiliado a la salud, 0 si está afiliado
}

alpha = 2.8 #ajustar segun sea lo necesario
theta = 2.5 #ajustar segun sea lo necesario
dim_t = len(umbrales) #numero de dimensiones
```

```python
# Calcular las brechas de pobreza y el índice BC 
def calcular_bc(df, umbrales, ponderadores, alpha, theta, dim_t):
    df['suma_brechas'] = 0
    for dim, umbral in umbrales.items():
        brecha = []
        for valor in df[dim]:
            if valor < umbral:
                b = (1 - valor / umbral)
            else:
                b = 0
            brecha.append(b)
        brecha = [(ponderadores[dim] / dim_t) * b ** theta for b in brecha]
        df[f'brecha_{dim}'] = brecha
        df['suma_brechas'] += df[f'brecha_{dim}']

    df['suma_brechas'] = [s ** (alpha / theta) if s != 0 else 0 for s in df['suma_brechas']]
    
    # Calcular el índice BC
    bc_index = df['suma_brechas'].sum() / len(df)
    return bc_index
```

```python
import matplotlib.pyplot as plt

# Suponiendo que los ponderadores y parámetros ya están definidos
ponderadores = {'p4010': 0.25, 'p6210s1': 0.25, 'inglabo': 0.25, 'p6090': 0.25}
alpha = 2
theta = 1
dim_t = 4

# Diccionario de DataFrames por departamento
dataframes_departamentos = {
    'Bogotá': df_bogotá,
    'Antioquia': df_antioquia,
    'ValleCauca': df_vallecauca,
    'Atlántico': df_atlántico,
    'Bolívar': df_bolívar
}

# Diccionario para almacenar los índices BC
bc_indices = {}

# Calcular el índice BC para cada departamento
for nombre, df in dataframes_departamentos.items():
    bc_index = calcular_bc(df, umbrales, ponderadores, alpha, theta, dim_t)
    bc_indices[nombre] = bc_index

for nombre, df in dataframes_departamentos.items():
    bc_index = calcular_bc(df, umbrales, ponderadores, alpha, theta, dim_t)
    print(f'Índice de Bourguignon y Chakravarty BC({alpha}, {theta}) para {nombre}: {bc_index:.4f}')

# Crear listas para los nombres de los departamentos y sus índices BC
departamentos = list(bc_indices.keys())
indices_bc = list(bc_indices.values())

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))  # Configurar el tamaño de la figura
plt.bar(departamentos, indices_bc, color='blue')  # Crear barras con los índices BC
plt.xlabel('Departamento')  # Etiqueta del eje X
plt.ylabel('Índice BC')  # Etiqueta del eje Y
plt.title(f'Índice de Bourguignon y Chakravarty BC({alpha}, {theta}) por Departamento')  # Título del gráfico
plt.xticks(rotation=45)  # Rotar las etiquetas del eje X para mejor visualización
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Añadir una cuadrícula horizontal
plt.tight_layout()  # Ajustar automáticamente los parámetros de la subtrama
plt.show()  # Mostrar el gráfico
```
